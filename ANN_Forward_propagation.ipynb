```python
import numpy as np

# Activation Function
def sigmoid(z):
    return 1 / (1 + np.exp(-z))


class SimpleANN:

    def __init__(self, n_input, n_hidden, n_output):

        # Input -> Hidden Layer
        self.W1 = np.random.randn(n_input, n_hidden) * 0.1 # array of random value (bec we dont know the real value)
        self.b1 = np.zeros((1, n_hidden))# bais array of 0s

        # Hidden -> Output Layer
        self.W2 = np.random.randn(n_hidden, n_output) * 0.1 # same as previous
        self.b2 = np.zeros((1, n_output))

    def forward(self, X):

        # Layer 1
        self.Z1 = np.dot(X, self.W1) + self.b1 # Z1 = XW1 + b1 (weighted sum for hidden layer)
        self.A1 = sigmoid(self.Z1) # calling the acivation function

        # Layer 2
        self.Z2 = np.dot(self.A1, self.W2) + self.b2 # # Z2 = A1 × W2 + b2
        self.A2 = sigmoid(self.Z2)

        return self.A2


# Input Data
X = np.array([
    [0,0],
    [0,1],
    [1,0],
    [1,1]
])

model = SimpleANN(2,4,1)

output = model.forward(X)

print(output)
```
