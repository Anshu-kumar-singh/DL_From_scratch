```python
import numpy as np

class Perceptron: # class is the blue print for creating the perceptron

    # in this we are deciding what should be the weight of the starting  values which are going to be used in this calculation
    def __init__(self, n_features, learning_rate=0.1, epochs=10): # this is called when we have to make a perceptron
        # start with weights = 0, bias = 0
        self.weights = np.zeros(n_features) # it mean w = [0.0 , 0.0 ]
        self.bias = 0
        self.lr = learning_rate
        self.epochs = epochs #  Train on all 4 samples 10 times bec one few times is not enough to get good result

    def activation(self, x): # it is relu activation function
        # step function: 1 if x >= 0 else 0
        return 1 if x >= 0 else 0

    def predict(self, x):
        z = np.dot(x, self.weights) + self.bias # z= (w^T) x + b (this is line formula)
        return self.activation(z)

    # The fit() method is the training algorithm of the perceptron.
    # Its job is to repeatedly look at every training example, compare the predicted output with the actual output, and adjust the weights and bias until the perceptron learns the correct decision boundary.
    def fit(self, X, y):
        for epoch in range(self.epochs):
            total_errors = 0

            for xi, target in zip(X, y): # zip() combines the input and its corresponding label.
                pred = self.predict(xi) # calling the function
                error = target - pred

                # update rule
                self.weights += self.lr * error * xi # updating the weight (wnew​=wold ​+ η × (y−y^​) × x)
                self.bias += self.lr * error # updatng the bais

                total_errors += int(error != 0)

            print(f"Epoch {epoch+1}: errors = {total_errors}")


# ---- Test it on AND gate ----
X = np.array([[0,0],[0,1],[1,0],[1,1]]) # input features (training data).
y = np.array([0,0,0,1]) # correct outputs (labels).

model = Perceptron(n_features=2)
model.fit(X, y)

for xi in X:
    print(xi, "->", model.predict(xi))
```
